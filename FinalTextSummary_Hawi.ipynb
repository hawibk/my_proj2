```python
import glob
import json
import random
import numpy as np
import pandas as pd
import re
import nltk

item = glob.glob(r'CORD-19-research-challenge/**/pdf_json/*.json',recursive=True)

#random_d = np.random.choice(item, int(len(item)*.1))

random_d = np.random.choice(item, int(len(item)*.0001))
```

```python
import os

file_list = []

for index, js in enumerate(random_d):
    with open(js) as json_file:
        json_text = json.load(json_file)
        
        for a in json_text['body_text']:
            
            only_text = a['text']
        
            file_list.append(only_text)
```

```python
import numpy as np

stop_words = nltk.corpus.stopwords.words('english')
stop_words
stop_words[:20]
len(stop_words)
```

```python
def normalize_document(txt):
    txt = re.sub(r'[^a-zA-Z0-9\s]', '', txt, re.I|re.A)
    txt = txt.lower()
    txt = txt.strip()
    tokens = nltk.word_tokenize(txt)
    clean_tokens = [t for t in tokens if t not in stop_words]
    return ' '.join(clean_tokens)

normalize_corpus = np.vectorize(normalize_document)
# New function that can take a list of items
norm_corpus = normalize_corpus(list(newFile))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()

transf = vect.fit_transform(newFile)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_y = cosine_similarity(transf[0:1], transf)
```

```python
t = cos_y * cos_y.T
```

```python
t
```

```python
import networkx as nx

nx_graph = nx.from_numpy_array(t)
scores = nx.pagerank(nx_graph)
```

```python
ranked_sentences = sorted(((scores[i],s) for i,s in enumerate(newFile)), reverse=True)
```

```python
for i in range(len(newFile)):
    print(ranked_sentences[i][1], '\n')
```
